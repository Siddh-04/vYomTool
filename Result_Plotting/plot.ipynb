```python
import uproot as up
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import awkward as ak
import scipy.stats as stats
# from

```

```python
event3_ele = up.open("Roo3.root")["myelectrons;1/Events;1"]
event4_ele = up.open("Roo4.root")["myelectrons;1/Events;1"]
event5_ele = up.open("Roo5.root")["myelectrons;1/Events;1"]

event3_mu = up.open("Roo3.root")["mymuons;1/Events;1"]
event4_mu = up.open("Roo4.root")["mymuons;1/Events;1"]
event5_mu = up.open("Roo5.root")["mymuons;1/Events;1"]
```

```python
ele3_pt = event3_ele["electron_pt"].array()
ele3_eta = event3_ele["electron_eta"].array()

ele3_pt_10 = ele3_pt[ele3_pt > 10]
# ele3_eta_10 = ele3_eta[ele3_pt > 10]

ele4_pt = event4_ele["electron_pt"].array()
ele4_eta = event4_ele["electron_eta"].array()

ele4_pt_10 = ele4_pt[ele4_pt > 10]
# ele4_eta_10 = ele4_eta[ele4_pt > 10]

ele5_pt = event5_ele["electron_pt"].array()
ele5_eta = event5_ele["electron_eta"].array()

ele5_pt_10 = ele5_pt[ele5_pt > 10]
# ele5_eta_10 = ele5_eta[ele5_pt > 10]

mu3_pt = event3_mu["muon_pt"].array()
mu3_eta = event3_mu["muon_eta"].array()

mu3_pt_10 = mu3_pt[mu3_pt < 10]
# mu3_eta_10 = mu3_eta[mu3_eta < 10]

mu4_pt = event4_mu["muon_pt"].array()
mu4_eta = event4_mu["muon_eta"].array()

mu4_pt_10 = mu4_pt[mu4_pt < 10]
# mu4_eta_10 = mu4_eta[mu4_eta < 10]

mu5_pt = event5_mu["muon_pt"].array()
mu5_eta = event5_mu["muon_eta"].array()

mu5_pt_10 = mu5_pt[mu5_pt < 10]
# mu5_eta_10 = mu5_eta[mu5_eta < 10]

```

```python
# flattening all arrays
ele3_pt_10 = ak.flatten(ele3_pt_10)
ele3_eta_10 = ak.flatten(ele3_eta)
ele4_pt_10 = ak.flatten(ele4_pt_10)
ele4_eta_10 = ak.flatten(ele4_eta)
ele5_pt_10 = ak.flatten(ele5_pt_10)
ele5_eta_10 = ak.flatten(ele5_eta)
mu3_pt_10 = ak.flatten(mu3_pt_10)
mu3_eta_10 = ak.flatten(mu3_eta)
mu4_pt_10 = ak.flatten(mu4_pt_10)
mu4_eta_10 = ak.flatten(mu4_eta)
mu5_pt_10 = ak.flatten(mu5_pt_10)
mu5_eta_10 = ak.flatten(mu5_eta)
```

```python
# plotting the histogram for each pt and eta
plt.figure()
plt.hist(ele3_pt_10, bins=100,label='ele3_pt_10',color='green',edgecolor='black',alpha=0.5)
plt.xlabel('electron pt')
plt.ylabel('number of electrons')
plt.title('electron transverse momentum for reading 3')
plt.legend()
plt.show()
```

```python
# similarly doing for eta
plt.figure()
plt.hist(ele3_eta_10, bins=100, label='ele3_eta_10',color='green',edgecolor='black',alpha=0.5)
plt.xlabel('electron eta')
plt.ylabel('number of electrons')
plt.title('electron eta for reading 3')
plt.legend()
plt.show()
```
