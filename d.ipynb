```python
import uproot as up
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import awkward as ak
import scipy.stats as stats
```

```python
event3_mu = up.open("Roo3.root")["mymuons;1/Events;1"]
```

```python
muon_pt = event3_mu["muon_pt"].array()
muon_eta = event3_mu["muon_eta"].array() # eta is the pseudorapiditiy
```

```python
fmuonpt=ak.flatten(muon_pt)
fmuoneta=ak.flatten(muon_eta)
```

```python
# plotting the histogram for muon pt only
plt.hist(fmuonpt, bins=100, range=(0,100), label='muon pt', color='lightgreen',edgecolor='black')
plt.xlabel('Muon pt')
plt.ylabel('Events')
plt.title('Muon pt for Roo3.root')
plt.legend()
plt.show()
```

```python
# similarly plotting for muon eta
plt.hist(fmuoneta, bins=100, range=(-5,5), label='muon eta', color='lightblue',edgecolor='black')
plt.xlabel('Muon eta')
plt.ylabel('Events')
plt.title('Muon  for Roo3.root')
plt.legend()
plt.show()
```

```python
# similarly getting data of elctrons eta and plotting it
event3_el = up.open("Roo3.root")["myelectrons;1/Events;1"]
electron_pt = event3_el["electron_pt"].array()
electron_eta = event3_el["electron_eta"].array() # eta is the pseudorapiditiy

felectronpt=ak.flatten(electron_pt)
felectroneta=ak.flatten(electron_eta)

# plotting the histogram for electron pt only
plt.hist(felectronpt, bins=100, range=(0,100), label='electron pt', color='lightgreen',edgecolor='black')
plt.xlabel('Electron pt')
plt.ylabel('Events')

plt.title('Electron pt for Roo3.root')
plt.legend()
plt.show()

# similarly plotting for electron eta
plt.hist(felectroneta, bins=100, range=(-5,5), label='electron eta', color='lightblue',edgecolor='black')
plt.xlabel('Electron eta')
plt.ylabel('Events')
plt.title('Electron eta for Roo3.root')
plt.legend()
plt.show()
```
